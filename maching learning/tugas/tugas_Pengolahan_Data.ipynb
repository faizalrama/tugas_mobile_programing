
# MEMASUKAN DATA

```python
data = pd.read_csv('data.csv')
```

```python
data
```

# TUGAS

## MENGHAPUS DATA KOSONG (nama,nim)

```python
data.dropna(subset=['nama', 'nilai'], inplace=True)
```

```python
data
```

## MERUBAH TIPE DATA (nilai(float) menjadi (Interger))

```python
data['nilai'].dtype
```

```python
data['nilai'] = data['nilai'].astype(int)
```

```python
data['nilai'].dtype
```

```python
data
```

## mengganti data alamat yang kosong menjadi saragi

```python
data['alamat'] = data['alamat'].fillna('saragi')
```

```python
data
```

## DATA ENCODING

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding untuk kolom 'alamat'
encoder = LabelEncoder()
data['alamat_encoded'] = encoder.fit_transform(data['alamat'])
```

```python
data[['alamat', 'alamat_encoded']].head()
```

```python
# One-hot encoding untuk kolom 'alamat'
data_encoded = pd.get_dummies(data, columns=['alamat'], prefix='alamat')

# Menampilkan hasil
data_encoded.head()
```

```python
# Target encoding untuk kolom 'alamat' berdasarkan rata-rata 'nilai'
data['alamat_target'] = data.groupby('alamat')['nilai'].transform('mean')

# Menampilkan hasil
data[['alamat', 'alamat_target']].head()
```

## pembagian data

### Split Berdasarkan Proporsi (Contoh: Train-Test Split)
Misalnya, 80% untuk data pelatihan dan 20% untuk data pengujian:

```python
from sklearn.model_selection import train_test_split

# Misalnya, membagi data 80-20
train, test = train_test_split(data, test_size=0.2, random_state=42)

print("Train set:")
print(train.head())

print("\nTest set:")
print(test.head())
```

```python
# Data dengan nilai >= 60
data_lulus = data[data['nilai'] >= 60]

# Data dengan nilai < 60
data_tidak_lulus = data[data['nilai'] < 60]

print("Data Lulus:")
print(data_lulus)

print("\nData Tidak Lulus:")
print(data_tidak_lulus)
```
